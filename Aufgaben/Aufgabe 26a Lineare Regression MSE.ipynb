# Übung: Lineare Regression und Mean Squared Error (MSE)

In dieser Übung arbeitest du mit einfachen Datenpunkten und einer Geraden.

Ziel:
1. Gegebene Datenpunkte visualisieren.
2. Eine Gerade zufällig initialisieren.
3. Vorhersagen der Geraden berechnen.
4. Den **Mean Squared Error (MSE)** zwischen Vorhersage und echten Werten berechnen.

Formel für den MSE:

$$MSE = \frac{1}{n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2$$

wobei:
- $y_i$ = echter Wert
- $\hat{y}_i$ = vorhergesagter Wert der Geraden
- $n$ = Anzahl der Datenpunkte


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Gegebene Datenpunkte

```python
# Datenpunkte
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([1.5, 2.1, 2.9, 3.8, 5.1, 5.3])

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gegebene Datenpunkte")
plt.show()
```

## Aufgabe 1: Zufällige Gerade initialisieren

Eine Gerade hat die Form:

$$ y = mx + b $$

- `m` = Steigung
- `b` = y-Achsenabschnitt

Initialisiere `m` und `b` zufällig.

```python
# Zufällige Parameter erzeugen
#m = 
#b = 

#print("m =", m)
#print("b =", b)

m = np.random.uniform(0, 2)
b = np.random.uniform(0, 1)

print("m =", m)
print("b =", b)
```

## Aufgabe 2: Vorhersagen berechnen

Berechne die vorhergesagten Werte der Geraden für alle `x`.

```python
#  Vorhersagen berechnen
#y_pred = 0
y_pred = m * x + b
print(y_pred)
y_pred
```

## Aufgabe 3: Mean Squared Error implementieren

```python
#  Implementiere eine Funktion für den MSE

def mse(y_true, y_pred):
    n = len(y_true)
    return np.sum((y_true - y_pred) ** 2) / n
error = mse(y, y_pred)
print("MSE:", error)
```

## Aufgabe 4: Visualisierung

Plotte die Datenpunkte und die zufällige Gerade.

```python
plt.scatter(x, y, label="Daten")
plt.plot(x, y_pred, label="Zufällige Gerade")
plt.legend()
plt.title("Daten vs. zufällige Gerade")
plt.show()
```

## Bonusaufgabe (optional)

- `m` und `b` manuell zu verändern
- zu beobachten, wie sich der **MSE** verändert.
